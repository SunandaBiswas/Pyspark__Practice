```
import pyspark
import matplotlib as plt
import seaborn as sns
```

```
vouchers = spark.table('bronze_clm.vouchers')
couponTypes = spark.table('bronze_clm.coupon_types')
```

```
display(vouchers)
display(couponTypes)
```

```
vouchers.printSchema() 
```

```
#creating dataframe using all the relevant columns from transaction table

vouchers_df = vouchers.select("id","coupon_type_id","account_id","customer_id", "transaction_id","discount_value","discount_type","create_date","modification_date")                                                                                                     
```

```
vouchers_df.limit(5).toPandas()
```

```
vouchers_df.printSchema()
```

```
from pyspark.sql import functions as F
voucherCreation = vouchers_df.select(F.date_format('create_date','MM').alias('month')).groupby('month').count().show()
```

```
vouchers_df.select("discount_value").distinct().count()
```

```
vouchers_df.groupBy("discount_value")
```

```
couponTypes.printSchema()
```

```
couponTypes_df = couponTypes.select("id", "modification_date", "code", "category", "value_type","ext_tag","ext_country", "ext_partner_id")   
```

```
couponTypes_df.limit(5).toPandas()
```

```
vouchers_couponTypes = couponTypes_df.join(vouchers_df,couponTypes_df.id == vouchers_df.coupon_type_id,"left")
```

```
vouchers_couponTypes.printSchema()
```

```
vouchers_couponTypes.select("vouchers.coupon_type_id").distinct().count()
```

```
couponTypes_df.select("coupon_types.id").distinct().count()
```

```
## types of discounts offered

vouchers_couponTypes.groupby('discount_type').count().orderBy('count', ascending=False).toPandas()
```

```
## several categories of coupons 

vouchers_couponTypes.groupby('category').count().orderBy('count', ascending=False).toPandas()
```
